```python
import boto3
import pandas as pd
```

```python
s3 = boto3.resource(
    service_name='s3',
    region_name='us-east-1'
)
```

```python
list_df = []
bucket = s3.Bucket('openaq-fetches')
objs = list(bucket.objects.filter(Prefix='realtime/2021-02-22/'))
for i in range(0, len(objs)):
    print(objs[i].key)
    s3file_obj = s3.Bucket('openaq-fetches').Object('realtime/2021-02-22/1613959266.ndjson').get()
    df = pd.read_json(s3file_obj['Body'], lines=True)
    print("New data shape:", df.shape)
    #print(df.head(2))
    list_df.append(df)
```

```python
len(list_df)
```

```python
combined_data = pd.concat(list_df)
```
